```python
import cv2

from matplotlib import pyplot as plt
import numpy as np
```

```python
cv2.__version__
```

```python
image = cv2.imread('flower.jpeg')
```

```python
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_rbg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

cv2.imshow('image _ 2' , image_rbg)
cv2.waitKey(0)
cv2.destroyAllWindows()
#we changed the red with blue
```

```python
# we can see the photo in inde
plt.imshow(image)
plt.axis("off")
plt.show()
#this change the place of rgb to bgr . 
#it is the reson of that show  (image)  with BGR
```

```python
plt.imshow(image_rbg)
plt.axis('off')
plt.show()
```

```python
def showimage (title = '' , image = None):
    plt.imshow(cv2.cvtColor(image , cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
showimage(title='image', image = image)
```
