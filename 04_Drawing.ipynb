```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
def imshowfunc(title = "" , image = None , size = 10) :
    h , w = image.shape[0] , image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = np.zeros((512,512,3) , np.uint8)
image_gray = np.zeros((512,512) , np.uint8)

imshowfunc("Black canvas - RGB" , image)
imshowfunc("Black canvas - GrayScale" , image_gray)
```

```python
cv2.line(image , (0,0) , (511,511) , (255,127,0) , 5)
imshowfunc("Diagonal Line" , image)
```

```python
image = np.zeros((512,512,3) , np.uint8)

cv2.rectangle(image , (100,100) , (300,250) , (127,50,127), 10)
imshowfunc('Rectangle',image)
```

```python
image = np.zeros((512,512,3) , np.uint8)

cv2.circle(image , (350,350), 100 ,(15,150,50) , -1)
#if the thickness be -1 it make a shape with full color 
imshowfunc('Circle' , image)
```

```python
image = np.zeros((512,512,3) , np.uint8)

image = np.zeros((512,512,3), np.uint8)

pts = np.array( [[10,50], [400,50], [90,200], [50,500]], np.int32)

pts = pts.reshape((-1,1,2))

cv2.polylines(image, [pts], True, (0,0,255), 3)
imshowfunc("Polygon", image)
```

```python
image = np.zeros((1000,1000,3), np.uint8)

text = 'Hello World!'

cv2.putText(image , text , (155,290) , cv2.FONT_HERSHEY_SCRIPT_SIMPLEX , 3 , (40,200,0) , 4)

imshowfunc('Text' , image)
```
