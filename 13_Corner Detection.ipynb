```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('tower.jpg')
imshow('Original' , image)

image2 = cv2.imread('chess.JPG')
imshow('Original' , image2)
```

```python
gray = cv2.cvtColor(image2 , cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)

haris_corners = cv2.cornerHarris(gray , 3 ,3 ,0.05) 

kernel = np.ones((7,7) , np.uint8)
haris_corners = cv2.dilate(haris_corners,kernel,iterations=2) 

image2[haris_corners > 0.025 * haris_corners.max()] = [255,127,127]
imshow('haris_corners' , image2)
```

```python
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)

haris_corners = cv2.cornerHarris(gray , 3 ,3 ,0.05) 

kernel = np.ones((7,7) , np.uint8)
haris_corners = cv2.dilate(haris_corners,kernel,iterations=2) 

image[haris_corners > 0.025 * haris_corners.max()] = [255,127,127]
imshow('haris_corners' , image)
```

```python
img = cv2.imread('chess.JPG')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(gray, 150, 0.0006, 10)

for corner in corners:
    x, y = corner[0]
    x = int(x)
    y = int(y)
    cv2.rectangle(img,(x-10,y-10),(x+10,y+10),(0,255,0), 2)
    
imshow("Corners Found", img)
```
